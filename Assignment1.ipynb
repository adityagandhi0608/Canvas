Question 2

```
first=input(' First Name')
last=input(' Last Name')
firs_list=list(first)
last_list=list(last)
firs_list.reverse()
last_list.reverse()
rev_first=''.join(str(x) for x in firs_list)
rev_last=''.join(str(x) for x in last_list)
print(f'{rev_first} {rev_last}')
```
